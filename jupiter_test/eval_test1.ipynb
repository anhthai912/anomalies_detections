```python
hello = get_eval.run_weights(
    iter_range= 50,
    iter_min_time= 30,
    mode= "train",
    predicted_path= "D:\\bi12year3\intern\gpu_slaves\\bau\\"
)
constant_hello = hello.copy()
sorted_hello = sorted(hello)
        
```

```python
min_val = min(hello)
print(min_val)

# min_val = [267.8124374296815, [195, 24]] #range= 300, #min_time =30
# [345.6901458731913, [50, 11]] range = 50, min time = 20
# [240.11465341759717, [280, 24]] maxed
```

```python
range_val = []
min_time_val = []
rmse_list = []

step = 100

for idx,i in enumerate(sorted_hello):
    if idx % step == 0:
        range_val.append(i[1][0])
        min_time_val.append(i[1][1])
        rmse_list.append(i[0])
```

```python
# Plot
plt.figure(figsize=(6, 16))
plt.plot(range_val, rmse_list)


plt.xlabel('Range')
plt.ylabel('RMSE')
plt.title('Plot of RMSE with Range')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(16, 6))
plt.plot(rmse_list, min_time_val)


plt.xlabel('RMSE')
plt.ylabel('min_time')
plt.title('Plot of RMSE with Min Time to detect')
plt.grid(True)
plt.show()
```
